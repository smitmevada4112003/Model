```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #sai kit learn 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("salary_integer_100.csv")
df.head()
```

```python
df.columns  # x= 2d y = 1d 
```

```python
X=df[['YearsExperience']] #input 2D
Y=df['Salary']
#1
print(X.head())
#print(Y.head()) #input and output  input - single value 
#  Degree  Age 
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(

    X,Y, test_size=0.2, random_state=42
)
```

```python
model=LinearRegression() #  modelconstructor  same linear class initalization 
```

```python
model.fit(X_train,Y_train)    #  input and out put 
```

```python
ypre=model.predict(X_test)
```

```python
r2 = r2_score(Y_test, ypre)
print("R2 Score :", r2)
```

```python
#y = mx + c  // 9449(2)+24k= y 
# output = slope (input) + constant 
print(model.coef_[0]) # experience 1 = 9k 
print(model.intercept_) # rapido= 50rs 
```

```python
result = model.predict([[6]])


print(result[0])
```

```python
plt.scatter(X,Y, color="blue")
plt.plot(X,model.predict(X),color="red")
plt.xlabel("experience")
plt.ylabel("salary")
plt.title("Expe vs salary")
plt.legend()

plt.show()
```
